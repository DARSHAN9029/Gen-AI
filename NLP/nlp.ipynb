# Stemming

```python
#1 Stemming

words=['eating','writing','eats','standing','taking','loving','programming','history','finalize']


#1 PorterStemmer

from nltk.stem import PorterStemmer
stemming=PorterStemmer()

for word in words:
    print(word+"----->"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

```python
#2 RegExStemmer Class
# Its basically takes a single regular expression and removes any prefix or suffix that mathches the expression.


from nltk.stem import RegexpStemmer
reg=RegexpStemmer('ing$|s$|e$|able$',min=4)
```

```python
reg.stem('eating')
```

```python
reg.stem('ingeating')
```

```python
# 3 Snowball STemmer
#better than porter stemmer

from nltk.stem import SnowballStemmer
sns=SnowballStemmer('english')
```

```python
for word in words:
    print(word+"---->"+sns.stem(word))
```

# Comapring Porter and Snowball Stemme

```python
stemming.stem("fairly"),stemming.stem("sportingly")
```

```python
sns.stem("fairly"),sns.stem("sportingly")
```

```python
sns.stem("goes")
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer

lematizer=WordNetLemmatizer()

lematizer.lemmatize("going",pos='v')
```

```python
for word in words:
    print(word+"---->"+lematizer.lemmatize(word,pos='v'))
```

```python
lematizer.lemmatize("goes",pos="v")
```

```python
lematizer.lemmatize("fairly")
```
